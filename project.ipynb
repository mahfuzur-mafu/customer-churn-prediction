```python
X= df[["Age","Gender","Tenure","MonthlyCharges"]]
y= df[["Churn"]]
```

```python
X
```

```python
y
```

Gender 1 = Female 
       0 = Male
Churn  1 = Yes
        2= No

```python
X["Gender"]= X["Gender"].apply(lambda x: 1 if x == "Female" else 0)
```

```python
X
```

```python
y["Churn"]= y["Churn"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
import joblib 
joblib.dump(scaler, 'scaler.pkl')
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy Score on model is {}".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
X
```

```python
y_pred = log_model.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid = {
    "n_neighbors" : [3, 5, 7, 9, 11, 13, 15],
    "weights" : ["uniform", "distance"]
}
```

```python
gridKn = GridSearchCV(KNeighborsClassifier(),param_grid, cv=5)
```

```python
gridKn.fit(X_train, y_train)
```

```python
gridKn.best_params_
```

```python
y_predkn =gridKn.predict(X_test)
```

```python
modelperformance(y_predkn)
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf','poly']
}
```

```python
gridsvc = GridSearchCV(svm, param_grid, cv=5)
```

```python
gridsvc.fit(X_train, y_train)
```

```python
gridsvc.best_params_
```

```python
y_predsvc= gridsvc.predict(X_test)
```

```python
modelperformance(y_predsvc)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
grid_tree =GridSearchCV(DecisionTreeClassifier(),param_grid, cv=5)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
y_preddt = grid_tree.predict(X_test)
```

```python
modelperformance(y_preddt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
param_grid = {
    "n_estimators": [32, 64, 128, 256],
    "max_features": [2, 3, 4],
    "bootstrap": [True, False]
}
```

```python
grid_rfc = GridSearchCV(rfc,param_grid, cv=5)
```

```python
grid_rfc.fit(X_train, y_train)
```

```python
grid_rfc.best_params_
```

```python
y_predrfc= grid_rfc.predict(X_test)
```

```python
modelperformance(y_predrfc)
```
